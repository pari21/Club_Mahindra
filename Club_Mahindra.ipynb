```python
# We will compare five different machine learning Regression models:

# 1 - Linear Regression
# 2 - Suport Vector Machine SVR
# 3 - Random Forest Regressor
# 4 = XGboost Regressor

# Function to calculate mean absolute error
def cross_val(model):
    from sklearn.model_selection import KFold
    from sklearn.metrics import mean_squared_error
    from sklearn.linear_model import LinearRegression
    from math import sqrt
    rmse = 0.0
    score = 0.0
    # k-fold technique
    i=1
    kf =  KFold(n_splits=5,random_state=0)

    for train_index, val_index in kf.split(train_data,label):
        print('\n{} of kfold {}'.format(i,kf.n_splits))
        xtr,xval = train_data.loc[train_index],train_data.loc[val_index]
        ytr,yval = label[train_index],label[val_index]

        model.fit(xtr,ytr)
        y_predicted = model.predict(xval)
        rmse = sqrt(mean_squared_error(yval, y_predicted))
        score = score + rmse
        print('rmse: ',rmse)
        i = i + 1

    print('final val score: ',score/5)
    


def predict_for_test(model):
    # Let's predict for test set
    test_predicted = model.predict(test_data)
    return test_predicted
```
